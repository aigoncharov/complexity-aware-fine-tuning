```python
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-8B")

think_end = tokenizer.encode("</think>")[0]
print("think end id", think_end)

print("think end seq", tokenizer.encode("</think>\n\n"))
```

```python
# Fix answer token index

import numpy as np

THINK_CHAIN_LIMIT = 5000


def fix_ans_token_idx(row):
    every_token_info = json.loads(row["qwen_8b_every_token_info"])
    chain_len = len(every_token_info)
    row["qwen_8b_ans_token_len"] = chain_len

    row["qwen_8b_ans_token_index"] = -1
    if chain_len < THINK_CHAIN_LIMIT:
        i = chain_len - 1
        # Check only last 10 tokens
        while i > 0 and i > chain_len - 10:
            token_probs_i = every_token_info[i]
            max_prob_i = np.argmax([token_prob["token_prob"] for token_prob in token_probs_i])
            if token_probs_i[max_prob_i]["token_idx"] == think_end:
                # Skip \n\n which is encoded as a single token 271
                row["qwen_8b_ans_token_index"] = i + 2
                break

            i -= 1

    return row


df_fixed = df.apply(fix_ans_token_idx, axis=1)
```

```python
# Verify we extracted answer token positions correctly by revalidating the answers

from reasoning_fine_tune.utils.correctness import check_answer_correct_mmlu


def check_ans_token_idx(row):
    ans_token_index = row["qwen_8b_ans_token_index"]

    if ans_token_index == -1:
        row["qwen_8b_ans_correct"] = False
        return row

    every_token_info = json.loads(row["qwen_8b_every_token_info"])

    ans_text = ""
    # Skip EOS token in the end
    for i in range(ans_token_index, len(every_token_info) - 1, 1):
        ans_token_probs = every_token_info[i]
        max_prob_i = np.argmax([token_prob["token_prob"] for token_prob in ans_token_probs])
        ans_token = ans_token_probs[max_prob_i]["token_idx"]
        ans_text += tokenizer.decode(ans_token)

    # print(row["answer_index"], ans_text)

    row["qwen_8b_extracted_answer_text"] = ans_text
    row["qwen_8b_ans_correct"] = check_answer_correct_mmlu(row, ans_text)

    return row


df_verified = df_fixed.apply(check_ans_token_idx, axis=1)
```
